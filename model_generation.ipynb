```python
# generative probability of the target words 
# ---------------------------------------------
import torch
import math
import numpy as np
from prettytable import PrettyTable

set_seed(42)

'''
More about Genearate: https://huggingface.co/docs/transformers/v4.27.2/en/main_classes/text_generation#transformers.GenerationMixin.generate

len(output.scores) = 10
len(outputs.sequences) = len(input) + 10

should we penalize the (rest) of the stop-words?
'''

# 1) first try with greedy search 
# generated_outputs = gpt2.generate(input_ids, top_k=50257, max_new_tokens=10, num_beams=1, do_sample=False, num_return_sequences=1, output_scores=True)
# generation =[tokenizer.decode(tok) for tok in generated_outputs.sequences]

# # logits to probs
# print(len(generated_outputs.scores))
# print(generated_outputs.scores[0].shape)
# probs = torch.stack(generated_outputs.scores, dim=1).softmax(-1) # probs.shape = torch.Size([3, 10, 50257])

# for syn in synonyms:
#   print(syn)
#   syn_ids = tokenizer(syn, return_tensors="pt").input_ids
#   print([probs[0, i,syn_ids[0, i].item()].item() for i in range(syn_ids.shape[1])])


# note: the softmax score for each token here are of power -17 and such, hence, it is not feasible to multiply them, or strong regularization is required.
# ---------------------------------------------

# 2) iterative confitional search through target words
# for syn more than 1 tokens (which is the most case), expand the input to input+t1 to see the conditional probability

# print(prompt)


def iterate_geneator(model_name, prompt, synonyms):

  model = AutoModelForCausalLM.from_pretrained(model_name, return_dict_in_generate=True)
  tokenizer = AutoTokenizer.from_pretrained(model_name)
  input_ids = tokenizer(prompt, return_tensors="pt").input_ids

  scores = []
  for syn in synonyms:
    # print(syn)
    syn_ids = tokenizer(syn, return_tensors="pt").input_ids
    conditional_scores = []

    generated_outputs = model.generate(input_ids, top_k=50257, max_new_tokens=1, num_beams=1, do_sample=False, num_return_sequences=1, output_scores=True)
    probs = torch.stack(generated_outputs.scores, dim=1).softmax(-1)
    # print('first-word distribution:', probs[0, 0, syn_ids[0, 0]].item())
    conditional_scores.append(probs[0, 0, syn_ids[0, 0]].item())

    if syn_ids.shape[1] > 1:
      for i in range(1, syn_ids.shape[1]+1): 
        toadd = torch.reshape(syn_ids[0,i-1], (1,-1))
        newinput_ids = torch.cat((input_ids, toadd), dim=1)
        generated_outputs = model.generate(newinput_ids, top_k=50257, max_new_tokens=1, num_beams=1, do_sample=False, num_return_sequences=1, output_scores=True)
        probs = torch.stack(generated_outputs.scores, dim=1).softmax(-1)
        # conditional_scores.append(probs[0, 0, syn_ids[0, i]].item())
        conditional_scores.append(math.log(probs[0, 0, syn_ids[0, i]].item()))

        if i+1==syn_ids.shape[1]:
          break

    # print(conditional_scores)
    # print(math.prod(conditional_scores))

    scores.append(conditional_scores)
  return scores 


scores_gpt = iterate_geneator("gpt2", prompt, synonyms=synonyms)
scores_biogpt = iterate_geneator("microsoft/biogpt", prompt, synonyms=synonyms)


myTable = PrettyTable(["Terms", "GPT AVG(LOG(SCORE))", "BioGPT AVG(LOG(SCORE))", "GPT Normalized SCORE", "BioGPT Normalized SCORE"])

for syn, score_gpt, score_biogpt in zip(synonyms, scores_gpt, scores_biogpt):
    # myTable.add_row([syn, math.prod(score_gpt), math.prod(score_biogpt)])
    myTable.add_row([syn, sum(score_gpt)/len(score_gpt), sum(score_biogpt)/len(score_biogpt)])
print(myTable)



# [!Archive]
# ---------------------------------------------

# first_word = probs[0,0,:]
# print("--------------------------------------")

# max word
# print("highest prob score: ", torch.max(first_word))
# print("highest prob id: ", torch.argmax(first_word))
# print("highest prob word: ", tokenizer.decode(torch.argmax(first_word)))
# print("--------------------------------------")

# top words
# top20 = torch.topk(first_word, 20)
# print(top20.indices)
# for i in top20.indices:
#   print(i)
#   print(tokenizer.decode(int(i)))
# print("--------------------------------------")


# for i in range(len(generated_outputs.scores)):
#     print(generated_outputs.scores[i])
#     print(generated_outputs.scores[i].shape[1])
#     print(torch.numel(generated_outputs.scores[i][generated_outputs.scores[i]!=float('-inf')])) 

# gpt2: among 50257 tokens, only 50 are possible (not -inf) of each position
# biogpt: among 42384 tokens, only 50 are possible (not -inf) of each position


# outputs = biogpt.generate(input_ids, top_k=25000, max_new_tokens=10, return_dict_in_generate=True, output_scores=True)
# transition_scores = model.compute_transition_scores(
#     outputs.sequences, outputs.scores, normalize_logits=False
# )

# # input_length is the length of the input prompt for decoder-only models, like the GPT family, and 1 for
# # encoder-decoder models, like BART or T5.
# input_length = 1 if model.config.is_encoder_decoder else inputs.input_ids.shape[1]
# generated_tokens = outputs.sequences[:, input_length:]

# print(transition_scores)
# print(generated_tokens)

# # for tok, score in zip(generated_tokens[0], transition_scores[0]):
# #     # | token | token string | logits | probability
# #     print(f"| {tok:5d} | {tokenizer.decode(tok):8s} | {score.numpy():.3f} | {np.exp(score.numpy()):.2%}")
# # print(transition_scores)
# # print(transition_scores[0])
# # print(transition_scores[0].shape)
```
